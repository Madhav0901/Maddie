```python
import pandas as pd
from scipy.stats import ttest_ind

# Load your dataset
data = pd.read_csv("C:/Users/madha/Desktop/scripts/bank_1.csv")

# Ensure 'deposit' column is properly formatted
data['deposit'] = data['deposit'].astype(str)  # Convert 'deposit' to string

# Ensure 'balance' column is numeric
data['balance'] = pd.to_numeric(data['balance'], errors='coerce')  # Convert 'balance' to numeric

# Handle missing or NaN values
data['balance'] = data['balance'].fillna(0)  # Replace NaN in 'balance' with 0
data = data.dropna(subset=['deposit'])  # Drop rows with NaN in 'deposit'

# Analyze Defaults: Group by 'deposit' and calculate mean for numerical columns
print("Group by 'deposit' and mean of numerical columns:")
print(data.groupby('deposit').mean())

# Perform T-test between groups ('Yes' and 'No')
yes_group = data[data['deposit'] == 'Yes']['balance']
no_group = data[data['deposit'] == 'No']['balance']

# T-test
t_stat, p_val = ttest_ind(yes_group, no_group, nan_policy='omit')
print(f"\nT-statistic: {t_stat}")
print(f"P-value: {p_val}")

# Interpret results
if p_val < 0.05:
    print("\nResult: The difference in balance between 'Yes' and 'No' groups is statistically significant.")
else:
    print("\nResult: No statistically significant difference in balance between 'Yes' and 'No' groups.")
```

```python
import pandas as pd
from scipy.stats import ttest_ind

# Load your dataset
# Replace 'data.csv' with your actual file path
data = pd.read_csv("C:/Users/madha/Desktop/scripts/bank_1.csv")

# Step 1: Inspect the dataset structure
print("Dataset Info:")
print(data.info())  # Check column types and null values
print("\nFirst 5 rows of the dataset:")
print(data.head())

# Step 2: Data Preparation
# Ensure 'deposit' column exists and is properly formatted
if 'deposit' not in data.columns:
    raise ValueError("The column 'deposit' does not exist in the dataset.")

data['deposit'] = data['deposit'].astype(str).str.strip()  # Convert 'deposit' to string and remove leading/trailing spaces

# Ensure 'balance' column exists and is numeric
if 'balance' not in data.columns:
    raise ValueError("The column 'balance' does not exist in the dataset.")

data['balance'] = pd.to_numeric(data['balance'], errors='coerce')  # Convert 'balance' to numeric

# Handle missing values in 'balance'
data['balance'] = data['balance'].fillna(0)  # Replace NaN in 'balance' with 0

# Step 3: Group Analysis
# Check unique values in 'deposit'
print("\nUnique values in 'deposit':", data['deposit'].unique())

# Select only numeric columns for groupby mean
numeric_columns = data.select_dtypes(include='number').columns
print("\nGroup by 'deposit' and mean of numerical columns:")
print(data.groupby('deposit')[numeric_columns].mean())

# Step 4: Perform T-test
# Separate groups based on 'deposit' values
yes_group = data[data['deposit'] == 'Yes']['balance']
no_group = data[data['deposit'] == 'No']['balance']

# Check if both groups have enough data
if yes_group.empty or no_group.empty:
    raise ValueError("One of the groups ('Yes' or 'No') has no data to perform a T-test.")

# Perform the T-test
t_stat, p_val = ttest_ind(yes_group, no_group, nan_policy='omit')

# Step 5: Output Results
print(f"\nT-statistic: {t_stat}")
print(f"P-value: {p_val}")

# Interpretation of the T-test result
if p_val < 0.05:
    print("\nResult: The difference in balance between 'Yes' and 'No' groups is statistically significant.")
else:
    print("\nResult: No statistically significant difference in balance between 'Yes' and 'No' groups.")
```

```python
import pandas as pd
from scipy.stats import ttest_ind

# Load your dataset
data = pd.read_csv("C:/Users/madha/Desktop/scripts/bank_1.csv")

# Step 1: Inspect the dataset structure
print("Dataset Info:")
print(data.info())  # Check column types and null values
print("\nFirst 5 rows of the dataset:")
print(data.head())

# Step 2: Data Preparation
# Ensure 'deposit' column exists and is properly formatted
if 'deposit' not in data.columns:
    raise ValueError("The column 'deposit' does not exist in the dataset.")

data['deposit'] = data['deposit'].astype(str).str.strip().str.capitalize()  # Normalize 'deposit'

# Ensure 'balance' column exists and is numeric
if 'balance' not in data.columns:
    raise ValueError("The column 'balance' does not exist in the dataset.")

data['balance'] = pd.to_numeric(data['balance'], errors='coerce')  # Convert 'balance' to numeric
data = data.dropna(subset=['balance'])  # Drop rows with NaN in 'balance'

# Step 3: Check Unique Values in 'deposit'
print("\nUnique values in 'deposit':", data['deposit'].unique())

# Step 4: Group Analysis
# Filter only rows with 'Yes' or 'No' in the 'deposit' column
valid_deposit_values = ['Yes', 'No']
data = data[data['deposit'].isin(valid_deposit_values)]

# Check if there is enough data for both groups
yes_group = data[data['deposit'] == 'Yes']['balance']
no_group = data[data['deposit'] == 'No']['balance']

if yes_group.empty or no_group.empty:
    print("\nError: One of the groups ('Yes' or 'No') has no data to perform a T-test.")
else:
    # Perform the T-test
    t_stat, p_val = ttest_ind(yes_group, no_group, nan_policy='omit')

    # Output T-test results
    print(f"\nT-statistic: {t_stat}")
    print(f"P-value: {p_val}")

    # Interpretation of the T-test result
    if p_val < 0.05:
        print("\nResult: The difference in balance between 'Yes' and 'No' groups is statistically significant.")
    else:
        print("\nResult: No statistically significant difference in balance between 'Yes' and 'No' groups.")

# Step 5: Final Dataset Summary
print("\nFinal processed data:")
print(data.head())
```

```python
import dash
from dash import dcc, html
from dash.dependencies import Input, Output
import pandas as pd
import plotly.express as px

# Load cleaned data
data = pd.read_csv("C:/Users/madha/Desktop/scripts/bank_1.csv")

# Initialize Dash app
app = dash.Dash(__name__)

# Layout of the dashboard
app.layout = html.Div([
    html.H1("Banking Data Dashboard", style={'textAlign': 'center'}),

    # Dropdown for selecting a segmentation column
    html.Label("Segment Data By:"),
    dcc.Dropdown(
        id='segment-dropdown',
        options=[
            {'label': 'Age Group', 'value': 'age'},
            {'label': 'Job', 'value': 'job'},
            {'label': 'Education', 'value': 'education'}
        ],
        value='age'  # Default value
    ),

    # Scatter plot for balance vs. duration
    dcc.Graph(id='scatter-plot'),

    # Bar chart for segment-based averages
    dcc.Graph(id='bar-chart')
])

# Callbacks for interactivity
@app.callback(
    [Output('scatter-plot', 'figure'),
     Output('bar-chart', 'figure')],
    [Input('segment-dropdown', 'value')]
)
def update_graphs(segment_by):
    # Scatter plot: Balance vs Duration
    scatter_fig = px.scatter(
        data, x='duration', y='balance',
        color=segment_by,
        title="Balance vs Duration",
        labels={'duration': 'Duration (Seconds)', 'balance': 'Balance (Amount)'}
    )

    # Bar chart: Average balance by segment
    bar_fig = px.bar(
        data.groupby(segment_by)['balance'].mean().reset_index(),
        x=segment_by, y='balance',
        title=f"Average Balance by {segment_by.capitalize()}",
        labels={'balance': 'Average Balance', segment_by: segment_by.capitalize()}
    )

    return scatter_fig, bar_fig

# Run the app
if __name__ == '__main__':
    app.run_server(debug=True)
```
